```

import math 
from matplotlib import pyplot as plt


Dimension = [2, 3, 4, 5, 10, 20, 30, 40, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]


def VolumenCorteza (dim, radio):
  # Volumen = radio * (1-0.99**dim)
  VolumenEsfera = ((math.pi**(dim/2)) / math.gamma((dim/2)+1)) * radio**dim
  VolumenCort = VolumenEsfera-(((math.pi**(dim/2)) / math.gamma((dim/2)+1)) * (0.99 * radio)**dim )
  return VolumenCort


def VolumenCortezaGiraud(dim, vol):
  VolumenCorteG = (vol)-(0.99 **dim)
  return VolumenCorteG

print(VolumenCorteza(Dimension[0], math.sqrt(1/math.pi)))
print(VolumenCortezaGiraud(Dimension[0], 1))

lista = []
for i in range (0, len(Dimension)-1):
  lista.append(VolumenCortezaGiraud(Dimension[i], 1))

print (lista)

plt.plot(lista)
plt.show

labels = ['a', 'b', 'c', 'd', 'e', '2','2','2','2','2','2','2','2','2']

for column, label in zip(lista, labels):
    plt.plot(column, label=label)
```

```
import math 
from matplotlib import pyplot as plt
import numpy as np

medio = [0,0]
ref = [1,0]

punto = []
puntos = []
puntos.append(ref)

angulo = 20
for i in range (1,18):
  punto = [math.cos(math.pi/(180/(i*angulo))),math.sin(math.pi/(180/(i*angulo)))]
  puntos.append(punto)

circ1 = []
circ2 = []
circ3 = []
circ4 = []

for j in range (len(puntos)):
  for k in range (j+1, len(puntos)):
    distancia = math.sqrt((puntos[k][0]-puntos[j][0])**2 + (puntos[k][1]-puntos[k][1])**2)
    circ1.append(distancia)
print("A: ", np.average(circ1))

# print(len(puntos))
fig, axs = plt.subplots(1,2)
axs[0].plot(*zip(*puntos), 'bo')

puntos.append(medio)
# print(len(puntos))

for j in range (len(puntos)):
  for k in range (j+1, len(puntos)):
    distancia = math.sqrt((puntos[k][0]-puntos[j][0])**2 + (puntos[k][1]-puntos[k][1])**2)
    circ2.append(distancia)
print("B: ", np.average(circ2))



axs[1].plot(*zip(*puntos), 'bo')

fig2, axs2 = plt.subplots(1,2)

puntos.remove(medio)
agregado = [1.5,0]

puntos.append(agregado)

for j in range (len(puntos)):
  for k in range (j+1, len(puntos)):
    distancia = math.sqrt((puntos[k][0]-puntos[j][0])**2 + (puntos[k][1]-puntos[k][1])**2)
    circ3.append(distancia)
print("C: ", np.average(circ3))

axs2[0].plot(*zip(*puntos), 'bo')

puntos.remove(ref)
puntos.remove(agregado)

for j in range (len(puntos)):
  for k in range (j+1, len(puntos)):
    distancia = math.sqrt((puntos[k][0]-puntos[j][0])**2 + (puntos[k][1]-puntos[k][1])**2)
    circ4.append(distancia)
print("D: ", np.average(circ4))
axs2[1].plot(*zip(*puntos), 'bo')



```
