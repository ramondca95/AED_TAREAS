```python
from random import randrange
import random
import matplotlib.pyplot as plt
iteration = 10

lista = []
lista2 = []
for j in range(3,10):
  win = 0
  contador = 0
  for i in range(0,iteration):

    cajas = [0] * j

    cajas[randrange(0,j)] = 1

    # print(cajas)

    decision1 = randrange(0,j)
    cajas.pop(decision1)

    cajas.remove(0)
   
    decision2 = random.choice(cajas)

    if decision2 == 1:
      win = win +1

  lista.append(win/iteration)
  lista2.append(j)
plt.plot(lista2,lista)






```
