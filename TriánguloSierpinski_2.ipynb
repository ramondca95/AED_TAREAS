```
import numpy as np
import matplotlib.pyplot as plt



# Cambiar a las iteraciones deseadas (del 1 al 7)
contador = 0
iterador = 2

# Puntos iniciales del triángulo antes de la primera iteración
init_point1 = (150,650)
init_point2 = (650,650)
init_point3 = (400,217)

# Arreglos que tendrá los componentes x y y de todos los puntos del triangulo
ejey = []
ejex = []

ejey.append(init_point1[1])
ejey.append(init_point2[1])
ejey.append(init_point3[1])

ejex.append(init_point1[0])
ejex.append(init_point2[0])
ejex.append(init_point3[0])

# Función para obtener el punto medio entre cada dos puntos
def midpoint(p1, p2):
    return((p1[0]+p2[0])/2, (p1[1]+p2[1])/2)

# Ciclo para la visualización de la actualización de la ventana donde se graficará el triángulo

# Fondo negro

# Se dibujan los puntos iniciales y se trazan las lineas del polígono que forman los tres puntos

points = [[init_point1,init_point2,init_point3]]

for k in range(1,iterador):
    contador += 1
    # Por si se quiere cerrar la ventana, que automáticamente finalice el programa

    array = []

    # Cada iteración de K aumenta los puntos que tendrá el arreglo points, por lo que irá haciendo más triángulos conforme K aumente.
    for point in points:
        l = []

        # Cada iteración trazará una linea del triángulo
        for i in range(0,3):
            #Se le da valor a las variables first y second, que contendrán los puntos medios entre dos puntos.
            first = midpoint(point[i-1],point[i])
            second = midpoint(point[i],point[(i+1)%3])


            #Se traza la linea entre uno y otro

            l += [[first,point[i],second]]

            # Las listas con los componentes en x y y van guardando de cada punto el valor correspondiente
            ejey.append(first[1])
            ejex.append(first[0])
            # print(points2[1])

            
        array += l

    points = array
# ejey.pop()
# print(ejey)
# print(ejey[0])

# ejex.pop()
# print(ejex)
# print(ejex[0])

listaEucl = []
listaManh = []
listaMink = []
listaCan = []

for dis in range (len(ejex)):

  for dist in range(dis+1,len(ejey)):
      # Fórmula de la distancia Euclideana
      euclideana = np.sqrt(((ejex[dis]-ejex[dist])**2)+((ejey[dis]-ejey[dist])**2))
      listaEucl.append(euclideana)
      listaEucl.sort()

      # Fórmula de la distancia Manhattan
      manhattan = np.abs(ejex[dis]-ejex[dist])+np.abs(ejey[dis]-ejey[dist])
      listaManh.append(manhattan)
      listaManh.sort()

      # Fórmula de la distancia Minkowski (Cambiar el valor de P al requerido)
      p = 1
      minkowski = ((np.abs(ejex[dis]-ejex[dist])**p) + (np.abs(ejey[dis]-ejey[dist])**p))**(1/p)
      listaMink.append(minkowski)
      listaMink.sort()
      # Fórmula de la distancia Canberra
      canberra = (np.abs(ejex[dis]-ejex[dist])+np.abs(ejey[dis] - ejey[dist]))/(np.abs(ejex[dis]))+(np.abs(ejex[dist]))+(np.abs(ejey[dis]))+(np.abs(ejey[dist]))
      listaCan.append(canberra)
      listaCan.sort()


# Para imprimir todas las distancias entre cada punto y los demás (úsese en casos con iteraciones menores a 2

print("\n******** Distancia Euclideana **********")

# print("Lista de distancias euclideanas")
# print(listaEucl)
print("Máxima distancia: ",np.max(listaEucl))
# print(np.max(listaEucl))
print("Mínima distancia: ", np.min(listaEucl))
# print(np.min(listaEucl))
print("Promedio de todas las distancias: ", np.mean(listaEucl))
# print(np.mean(listaEucl))
plt.plot(listaEucl)
plt.title("Euclideanacon %i iteraciones" %iterador)
plt.show()

print("\n******** Distancia Manhattan **********")
# print("\nLista de distancias Manhattan")
# print(listaManh)
print("Máxima distancia: ", np.max(listaManh))
print("Mínima distancia: ", np.min(listaManh))
print("Promedio de todas las distancias: ", np.mean(listaManh))
plt.plot(listaManh)
plt.title("Manhattan con %i iteraciones" %iterador)
plt.show()

print("\n******** Distancia Minkowski **********")
print("Para cuando P vale: ", p)
# print("\nLista de distancias Minkowski")
# print(listaMink)
print("Máxima distancia: ", np.max(listaMink))
print("Mínima distancia: ", np.min(listaMink))
print("Promedio de todas las distancias: ", np.mean(listaMink))
plt.plot(listaMink)
plt.title('Minkowski con P = '+ str(p)+' y %i iteraciones' %iterador)
plt.show()

print("\n******** Distancia Canberra **********")
# print("\nLista de distancias Canberra")
# print(listaCan)
print("Máxima distancia: ", np.max(listaCan))
print("Mínima distancia: ", np.min(listaCan))
print("Promedio de todas las distancias: ", np.mean(listaCan))
plt.plot(listaCan)
plt.title("Canberracon %i iteraciones" %iterador)
plt.show()
```
